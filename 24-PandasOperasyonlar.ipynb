Group By  sıralama

```python
maasSozlugu = {"Departmanlar" : ["Yazılım","Yazılım","Satış","Satış","Pazarlama","Pazarlama","Finans","Finans","İnsan Kaynakları","İnsan Kaynakları","Hukuk","Hukuk"],
               "Çalışanlar" : ["Asım Yılmaz","Haluk kar","Atakan Kemal","Yusuf çınar","Hasan Hüseyin","Kamil er","Canan Can","Seda Canan","Sertaç Canan","Talat ataş","Ali Can","Can Can"],
               "Maaşlar" : [100,150,200,200,300,400,500,250,450,300,200,750]}
```

```python
maasSozlugu = pd.DataFrame(maasSozlugu)
```

```python
maasSozlugu
```

```python
grupObjesi= maasSozlugu.groupby("Departmanlar")
```

```python
grupObjesi.count() #departmanlara göre gruplar ve sayar 
```

```python
grupObjesi.mean() #departmanlara göre gruplar ve ortalamasını alır
```

```python
grupObjesi.min() #departmanlara göre gruplar ve min değerini alır
```

```python
grupObjesi.max() #departmanlara göre gruplar ve max değerini alır
```

```python
grupObjesi.describe() #departmanlara göre gruplar ve istatistiksel bilgileri verir
```
