```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
(benimFigur,benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#c96f23",alpha=0.5) #alpha saydamlık
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="blue",alpha=0.5)
#rgb hex code araştır. #3a4f5c gibi
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,color="blue",linewidth=3.0) #linewidth çizgi kalınlığı
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,color="yellow",linewidth=3.0) #linewidth çizgi kalınlığı
yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,color="black",linestyle="--") #linestyle çizgi stili
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,color="green",linestyle=":") #linestyle çizgi stili
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,color="#000000",linestyle="-.",marker="o",markersize=4,markerfacecolor="red") #linestyle çizgi stili
yeniEksen.plot(numpyDizisi1,numpyDizisi1+7,color="#000000",linestyle="-.",marker="o",markersize=4,markerfacecolor="yellow",markeredgewidth=3,markeredgecolor="blue") #linestyle çizgi stili
```

# scatter
