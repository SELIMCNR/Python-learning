```python
y = dataFrame["Type"].values #type özelliğini y değişkenine atar.
x = dataFrame.drop("Type",axis=1).values #type özelliğini hariç tutarak x değişkenine atar.
```

```python
from sklearn.model_selection import train_test_split #veri setini train ve test olarak ayırmak için
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15) #veri setini train ve test olarak ayırır.
```

```python
from sklearn.preprocessing import MinMaxScaler #veri setini ölçeklendirmek için
```

```python
scaler = MinMaxScaler() #ölçeklendirme için nesne oluşturur.
```

```python
scaler.fit(x_train) #ölçeklendirme için x_train verisini kullanır.
```

```python
x_train = scaler.transform(x_train) #x_train verisini ölçeklendirir.
```

```python
x_test = scaler.transform(x_test) #x_test verisini ölçeklendirir.
```

```python
import tensorflow as tf #tensorflow kütüphanesini kullanmak için
from keras.models import Sequential #model oluşturmak için
from keras.layers import Dense #katman oluşturmak için 
from keras.layers import Dropout #overfittingi engellemek için
from keras.layers import Activation #aktivasyon fonksiyonu eklemek için
from keras.callbacks import EarlyStopping #overfittingi engellemek için
```

```python
x_train.shape #x_train verisinin boyutunu gösterir.
```

```python
model = Sequential() #model oluşturur.
```

```python
model.add(Dense(units=30,activation="relu")) #30 nöronlu gizli katman ekler.
model.add(Dense(units=15,activation="relu")) #15 nöronlu gizli katman ekler.
model.add(Dense(units=15,activation="relu")) #15 nöronlu gizli katman ekler.
model.add(Dense(units=1,activation="sigmoid")) #1 nöronlu çıkış katmanı ekler. sigmoid aktivasyon fonksiyonu kullanır.

model.compile(loss="binary_crossentropy",optimizer="adam") #modeli derler.
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1) #modeli eğitir. #epochs: kaç kere eğitileceği, validation_data: test verisi, verbose: eğitim sırasında bilgi verir.  
```

```python
modelKaybi = pd.DataFrame(model.history.history) #modelin eğitim kaybını alır.
```

```python
modelKaybi.plot() #modelin eğitim kaybını grafikle gösterir.
```

## overfittingi engellemek için early stopping kullanılır. 

```python
model.add(Dense(units=30,activation="relu")) #30 nöronlu gizli katman ekler.
model.add(Dense(units=15,activation="relu")) #15 nöronlu gizli katman ekler.
model.add(Dense(units=15,activation="relu")) #15 nöronlu gizli katman ekler.
model.add(Dense(units=1,activation="sigmoid")) #1 nöronlu çıkış katmanı ekler. sigmoid aktivasyon fonksiyonu kullanır.

model.compile(loss="binary_crossentropy",optimizer="adam") #modeli derler.
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25) #overfittingi engellemek için , monitor: hangi metriğe göre, mode: min veya max da tutma verbosu, verbose: bilgi verir, patience: kaç epoch sabredeceğiz.
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping]) #modeli eğitir. #epochs: kaç kere eğitileceği, validation_data: test verisi, verbose: eğitim sırasında bilgi verir, callbacks: overfittingi engellemek için.
```

```python
modelKaybi = pd.DataFrame(model.history.history) #modelin eğitim kaybını alır.
```

```python
modelKaybi.plot() #modelin eğitim kaybını grafikle gösterir.
```

```python
#Dropout: öğrenme sırasında rastgele nöronları kapatır. Overfittingi engeller.
```

```python
model.add(Dense(units=30,activation="relu")) #30 nöronlu gizli katman ekler.
model.add(Dropout(0.6)) #Dropout ekler. 0.6 oranında nöronları kapatır.
model.add(Dense(units=15,activation="relu")) #15 nöronlu gizli katman ekler.
model.add(Dropout(0.6)) #Dropout ekler. 0.6 oranında nöronları kapatır.
model.add(Dense(units=15,activation="relu")) #15 nöronlu gizli katman ekler.
model.add(Dropout(0.6)) #Dropout ekler. 0.6 oranında nöronları kapatır.
model.add(Dense(units=1,activation="sigmoid")) #1 nöronlu çıkış katmanı ekler. sigmoid aktivasyon fonksiyonu kullanır.

model.compile(loss="binary_crossentropy",optimizer="adam") #modeli derler.
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping]) #modeli eğitir. #epochs: kaç kere eğitileceği, validation_data: test verisi, verbose: eğitim sırasında bilgi verir, callbacks: overfittingi engellemek için.
```

```python
kayipDf = pd.DataFrame(model.history.history) #modelin eğitim kaybını alır.
```

```python
kayipDf.plot() #modelin eğitim kaybını grafikle gösterir.
```

```python
predict_x = (model.predict(x_test) > 0.5).astype("int32") #modelin tahminlerini alır. ikili sınıflandırma olduğu için 0.5 ten büyükse 1, küçükse 0 yapar.
#tahminlerimiz= np.argmax(predict_x,axis=1) #np.argmax fonksiyonu ile tahminlerimizi alırız. çok sınıflı sınıflandırma olduğu için axis=1 veririz.
tahminlerimiz= predict_x    
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report,confusion_matrix #modelin başarısını ölçmek için
```

```python
print( classification_report(y_test,tahminlerimiz)) #modelin başarısını ölçer. #precision: kesinlik, recall: duyarlılık, f1-score: harmonik ortalama, support: destek
```

```python
print( confusion_matrix(y_test,tahminlerimiz)) #modelin başarısını ölçer. #hata matrisi ,7 yanlış tahmin etmiş 10 diğerleri doğru.
```
