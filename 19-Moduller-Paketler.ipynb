```python
import numpy as np   #numpy kütüphanesi eklendi modülü eklendi np adında kullanacağız 
import matplotlib.pyplot  as matplot  #matplotlib kütüphanesi eklendi matplot adında
```

```python
#numpy verileri oluşturu
maasListesi = np.random.normal(4000,500,1000) #normal dağılımda random rastgele(4000 lira civarında,500 lira standart sapmalı,1000 tane veri oluştur)
np.mean(maasListesi)  #listenin ortalamasını al
```

```python
#matplotlib verileri görselleştirir.
matplot.hist(maasListesi,50)   #bu listeden 50 veriyi histogram görüntüde görselleştir
matplot.show                   #tabloyu göster
```
