```python
import numpy as np
import matplotlib.pyplot as plt

#matplotlib inline komutu ile grafiklerin notebook üzerinde görünmesini sağlıyoruz.
%matplotlib inline 
```

```python
yasListesi=[11,20,39,42,57,65,70,75]
```

```python
kiloListesi=[20,40,74,83,69,72,86,71]
```

```python
numpyYasListesi=np.array(yasListesi)
```

```python
numpykiloListesi=np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpykiloListesi
```

```python
plt.plot(numpyYasListesi,numpykiloListesi,"g") #(x ekseni,y ekseni,renk) grafik oluştur.
# r : red ,g : green , b : blue , c : cyan , m : magenta , y : yellow , k : black , w : white
plt.xlabel("Yas") #x ekseni ismi
plt.ylabel("Kilo") #y ekseni ismi
plt.title("Kilonun Yaşa Göre Değişimi") #grafik başlığı
plt.show() #grafik çizdirme normalde bu satırı yazmamıza gerek yok. ama bazı durumlarda yazmamız gerekiyor.
```

```python
numpyDizisi1 = np.linspace(0,20,50) #0 ile 20 arasında 20 tane sayı üret.   
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3 #numpyDizisi1 in küplerini al.
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--") #g-- : green : yeşil renk çizgi grafik oluştur. g*- : green : yeşil renk yıldız grafik oluştur.
```

```python
plt.subplot(1,2,1) #1 satır 2 sütunluk bir grafik oluştur ve 1.grafikte çalış demek.
plt.plot(numpyDizisi1,numpyDizisi2,"r*-") #1.grafik
plt.subplot(1,2,2) #1 satır 2 sütunluk bir grafik oluştur ve 2.grafikte çalış demek.
plt.plot(numpyDizisi2,numpyDizisi1,"g--") #2.grafik
```

```python
benimFigure = plt.figure() #boş bir figure oluştur.
figureAxes = benimFigure.add_axes([0.1,0.1,0.3,0.3]) #figure içerisinde axes oluştur. axes : grafik çizilebilecek alan. [sol,alt,genişlik,yükseklik] 0 ile 1 arasında değer alır.
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi") #x ekseni ismi
figureAxes.set_ylabel("Y ekseni veri ismi") #y ekseni ismi
figureAxes.set_title("Grafik başlığı") #grafik başlığı
```

```python
benimFigure 
```

```python
figure2 = plt.figure() #boş bir figure oluştur.
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) #figure içerisinde axes oluştur. axes : grafik çizilebilecek alan. [sol,alt,genişlik,yükseklik] 0 ile 1 arasında değer alır.
eksen2 = figure2.add_axes([0.3,0.5,0.3,0.3]) #figure içerisinde axes oluştur. axes : grafik çizilebilecek alan. [sol,alt,genişlik,yükseklik] 0 ile 1 arasında değer alır.
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni veri ismi") #x ekseni ismi
eksen1.set_ylabel("Y ekseni veri ismi") #y ekseni ismi
eksen1.set_title("Ana Grafik başlığı") #grafik başlığı

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X ekseni veri ismi küçük") #x ekseni ismi
eksen2.set_ylabel("Y ekseni veri ismi küçük") #y ekseni ismi
eksen2.set_title("Küçük Grafik başlığı") #grafik başlığı
```

```python
(benimFigure,benimEksenler)=plt.subplots(nrows=1,ncols=2) #1 satır 2 sütunluk bir figure oluştur. benimFigure : figure , benimEksen : axes
for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni veri ismi") #x ekseni ismi
plt.tight_layout() #grafiklerin birbirine girmemesi için kullanılır.
```

```python
yeniFigur = plt.figure(figsize=(5,5),dpi=1000)
#figsize : figure boyutu (genişlik,yükseklik) inch cinsinden çok değiştirmeye gerek yok duruma göre değiştir.
#dpi : dot per inch (inç başına düşen piksel sayısı) kayıt ederken dpi gir kaliteli gözüksün grafik.
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi2 ** 2,label="numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi2 ** 3,label="numpyDizisi ** 3")
yeniEksen.legend(loc=2) #loc=2 : 2.konumda legend oluştur. loc=0 : en uygun konumda legend oluştur.
```

```python
yeniFigur.savefig("benimfigur.png",dpi=1000) #oluşturulan figure kayıt etme.
```
